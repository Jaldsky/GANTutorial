## Основные объекты и понятия

Основу <b>Pandas</b> составляют такие структуры данных, как <b>Series и DataFrame</b>.

<b>Series</b> - это <b>одномерный массив</b>, имеющий специальные <b>метки (индексы)</b> и способные хранить даннеы <b>любого</b> типа. 

<b>DataFrame</b> - это <b>двумерный массив (матрица, таблица)</b>, имеющий <b>специальные метки (индексы)</b>, который хранит в своих столбцах данные <b>разных типов</b>.

## Series

Основы

```python
s = pd.Series(["Pandas", "Matplotlib", "Numpy", np.nan]) # простая серия
s
```

```python
s = pd.Series(np.random.randn(4), index=['a', 'b', 'c', 'd']) # серия с подписью индексов
print(s)
print(s.index) # вывод столбца с индексом

print(s.dtype) # тип значений в словаре
print(s.shape) # форма
print(s.size) # размер
print(s.ndim) # колиество измерений
```

```python
n = {'q': 2, 'p': 2, 'r': 4} # словарь dict
pd.Series(n) # создание серии из dict
```

Перевод Series/DataFrame к Array numpy

```python
s.to_numpy() # перевод Series в Array numpy
```

## Индексы и выборка

Обращение по индексам

```python
print(s, '\n') # DataFrame 


print(s[0], '\n') # обращение к первому элементу


print(s[1:5], '\n') # обращение к элементам от 1 до 5

# обращение по условию
print(s[s > 0], '\n') 
print(s[(s > 0) & (s < 1)], '\n')

# обращение по индексу
print(s['c'], '\n') 
print(s[['c','d']])
```

## Добавление и удаление

Добавление элементов

```python
s['py'] = 99 # содание новой строки с именем py и значением 99
s
```

Удаление элементов

```python
s.drop(labels=['a', 'py']) # удаление по индексу, также можно использзовать, index, axis и т. д.
```

## Операции

Подсет синуса, тангена, суммы, среднего значения, максимума, модуля

```python
print(np.sin(s), '\n') # подсчет синуса
print(np.tan(s), '\n') # подсчет тангенса
print(np.abs(s), '\n') # подсчет модуля

print(s.sum(), '\n') # подсчет суммы
print(s.mean(), '\n') # подсчет среднего значения
print(s.max()) # максимальное значение
```

Стандартные операции

```python
print(s + s) # сумма матриц
print(s * s) # произведение матриц
print(s ** 2) # возведениие в квадрат
```

Перевод из одного типа в другой

```python
s.astype(np.int16) # перевод из float64 в int16
```

Доступ к строкам по индексу возможен несколькими способами:

 *   .loc - используется для доступа к значению по строковой метке (индексу) (by row & column index label)
 *   .iloc - используется для доступа по числовому значению (начиная от 0) (by row & column number)
 
 .at и .iat анаологичны, но могут получать только одно значение за раз
 
 ix - устарел


```python
df.loc['KZ'] # at
```

```python
df.loc[['KZ'], ['square']] # строка с меткой KZ, стобец с меткой square
```

```python
df.iloc[1] # 1 строка, все стобцы , iat
```

```python
df.iloc[1, 1] # 1 строка и 1 столбец
```

```python
df.iloc[1, 1:3] # 1 строка c 1 по 3 столбец
```

```python
df.iloc[1:3, 1:3] # c 1 по 3 строки и c 1 по 3 столбец
```

```python
df.loc[['KZ', 'RU'], 'square'] # по индексу и интересующим колонкам
```

```python
df.loc['RU':'UA', :] # срезы от RU до UA со всеми столбцами :
```

## Группировка и сортировка 

Сортировка значений

```python
df.sort_values('Age',ascending=False).head(10) #sort_index сортировка возроста по убыванию
```

```python
df.groupby(['Sex']).count() # группировка по полу и подсчет
```

```python
df.groupby(['Sex'])['Age'].agg(np.mean) # группировка по полу, вычисление среднего значения по возрасту
```

```python
pvt = df.pivot_table(index=['Sex'], columns=['Pclass'], values='Name', aggfunc='count')
pvt # сводная таблица сколько всего женщин и мужчин было в конкретном классе корабля
```

Связывание с помощью join

```python
n_df1 = pd.DataFrame({
'country': ['Kazakhstan', 'Russia', 'Belarus', 'Ukraine','Kazakhstan'],
'population': [17.04, 143.5, 9.5, 45.5, 232.12],
'square': [2724902, 17125191, 207600, 603628, 35445]
}, index=['KZ', 'RU', 'BY', 'UA', 'KZ'])

n_df2 = pd.DataFrame({
'country': ['Kazakhstan', 'Russia', 'Belarus', 'Ukraine','Kazakhstan'],
'some_values': [542.03, 657.5, 6.5, 54.5, 543.15],
'more_values': [423423, 423423, 65464, 4342, 656]
}, index=['GE', 'RU', 'FR', 'IT', 'US'])
```

```python
# n_df1.join(n_df2, on='country',how='right', lsuffix='_right', rsuffix='_left') 
# SQL-style объединяет столбцы df1 и df2: 'left', 'right', 'outer', 'inner'

pd.merge(n_df1, n_df2,how='left', on='country')
```
