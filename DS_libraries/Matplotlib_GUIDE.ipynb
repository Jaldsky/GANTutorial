## <i>Рисунки и графики. Основная идея matplotlib</i>

**Matplotlib** – библиотека для визуализации данных двумерной (2D) графикой.
<br>**Рисунок(figure)** - общее окно графика, в нем можно построить один или несколько графиков.
<br>**Оси(axes)** - сам график, с осями и легендой.

image.png

```python
# %matplotlib inline #график выводится в jupyter notebook как обычная картинка
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # стиль графика
```

```python
#построение графика
x1 = [1, 2, 4, 6, 6, 7, 9]
y1 = [1, 3, 5, 6, 6, 7, 9]

x2 = [0, 5, 5, 7, 7, 8, 9]
y2 = [2, 4, 6, 7, 7, 8, 9]


plt.plot(x1, y1)
```

```python
figures, axes = plt.subplots(2, 2) #создание матрицы 2х2 с рисункоми(figures) и осями(axes)
# figures.add_subplot(2, 2, 3)
axes[0, 0].plot(x1, y1) #отрисовка графика в матрице, [0, 0] - позиция графика
axes[1, 1].plot(x2, y2)
axes[0, 1].plot (x1, y1, x2, y2)
```

```python
axes # матрица из 4-х графиков
```

```python
# добавляем графики по отдельности 
figure = plt.figure()

# figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None)
# num - номер рисунка, figsize - (float, float) ширина и высота в дюймах, dpi - разрешение, facecolor -цвет фона,
# edgecolor - цвет краёв 

ax1 = figure.add_subplot(2, 2, 1) # расчервиваем рисунок на 4 части и вставляем график в 1 ячейку 
ax2 = figure.add_subplot(2, 2, 2) # вставляем график во 2 ячейку 
ax3 = figure.add_subplot(2, 2, 3) # вставляем график в 3 ячейку 
# 4 соотвественно пустая 
ax1.plot(x1, y1)
ax3.plot(x2, y2)
```

## Цвет линий, стили и маркера


Цвет и стили линий

```python
import numpy as np

fig, ax = plt.subplots() #график и оси

#входные данные
x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

#построение с различными цветами и стилями линий
ax.plot(x, y)

ax.plot(x, np.sin(x), color=(1.0,0.2,0.3),linestyle='-')

ax.plot(x, x + 5, color='blue', linestyle='--')

ax.plot(x, x + 3, color='g',linestyle=':')    

ax.plot(x, np.cos(x), color='0.75',linestyle='-.')

ax.plot(x, x, color='#FFDD44',linestyle='--')
```

Маркера

```python
fig, ax = plt.subplots() #график и оси

#входные данные
x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

#построение с различными маркерами
ax.plot(x, y, marker='o')

ax.plot(0.5 * x, y, marker='^')

ax.plot(2 * x, y, marker='*')
```

```python
fig, ax = plt.subplots() #график и оси

#входные данные
x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='--', marker='o', alpha=0.6) # alpha=0.2 - прозрачность
# c='red', ls='--' сокращения о color и linestyle
ax.plot(0.5 * x, y,'*--g', linewidth=2.5) # толщина линии linewidth
# '*--g' сразву совмещаем парамметры типл линии ls='--' и c='g' от green
```

## Лимиты осей

```python
fig, ax = plt.subplots() #график и оси

#входные данные
x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

#построение графика
ax.plot(x, y, color='red', linestyle='-', marker='o')
#ограничиваем 
ax.set_xlim(0, 11) # x может принять значения от 0 до 11
ax.set_ylim(-5, 10) # y может принять значения от -5 до 10
```

## Подписи и сетка гафика

Настройка подписей

```python
fig, ax = plt.subplots() #график и оси

#входные данные
x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

#построение графиков
ax.plot(x, y, color='red', linestyle='-', marker='o', label='y=x^2') #label - метка(подпись) к графику
 
ax.plot(0.5 * x, y, color='green', linestyle='-', marker='*', label='y=0.5 * x^2')

ax.set_title('Заголовок нашего графика', fontsize=20) #заголовок к графику

#метки к осям
ax.set_xlabel("Ось х") 
ax.set_ylabel("Ось у")

ax.legend(loc='lower left') #расположение легенды в нашем случае меток к графикам (labels)

# ax.cla() #очищаем график
# plt.clf()  #очистить рисунок 
```

Настройка сетки

```python
fig, ax = plt.subplots() #график и оси

#входные данные
x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

#построение графика
ax.plot(x, y, color='red', linestyle='-', marker='o', label='y=x^2')

#настройка решетки графика
ax.set_xticks(np.arange(-3, 3, 0.5)) #от -3 до 3 с шагом 0.5
ax.set_yticks(np.arange(0, 10, 3)) #от 0 до 10 с шагом 3

ax.spines['right'].set_visible(False) #убираем серую линию справа  
ax.spines['left'].set_visible(False) #убираем серую линию слева  

ax.grid(color = 'blue', linewidth = 2, linestyle = '--') 
# чтобы убрать сетку полностью ax.grid()
```

## Виды графиков

График рассеивания (scatter)

```python
# scatter plot

fig, ax = plt.subplots() #график и оси

#входные данные
N = 50
x = np.random.rand(N)
y = np.random.rand(N)


colors = np.random.rand(N) #выбрать уникальный цвет для каждой точки
area = (30 * np.random.rand(N))**2 #площадь

#построение графика
ax.scatter(x, y, s=area, c=colors, alpha=0.8, marker='.', cmap='bwr') #cmap='viridis' - цветовая схема
# цветовые схемы можно посмотреть тут https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

3D график рассеивания (scatter)

```python
# 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(projection='3d') #указываем проекцию 3D
#построение графика
ax.scatter(x, y, s=area, c=colors, alpha=0.5, marker='.', cmap='bwr')
```

График баров (bars)

```python
#bars

fig, axes = plt.subplots(1, 2) #график и оси

#построение графиков
axes[0].bar([1,2,3],[3,4,5], color='red',  label='vertical bar', alpha=0.8)#bar вертикальное расположение
axes[1].barh([0.5,1,2.5],[0,1,2]) #barh горизонтальное расположение

#отображение легенды
axes[0].legend()
```

Столбчатая диаграмма с накоплением (Stacked bar chart)

```python
# Stacked bar chart

fig = plt.figure(figsize=(14,10))

#построение графиков
rect1 = plt.bar(np.arange(5), np.arange(5)**2,
                width=0.5, color='lightblue')

rect2 = plt.bar(np.arange(5),  np.arange(5) * 2,
                width=0.5, color='#1f77b4')
#отображение легенды
plt.legend((rect1[0], rect2[0]), ('Support', 'Freedom'))
```

Hist и boxplot

```python
# hist и boxplot

fig, ax = plt.subplots() #график и оси

np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

ax.hist(x, 50, density=True, facecolor='g', alpha=0.75) #density - плотность
```

Японские свечи (boxplot)

```python
#входные данные
user_1 = [10, 3, 15, 21, 17, 14]
user_2 = [5, 13, 10, 7, 9, 12]
data = [user_1, user_2]

fig = plt.figure(figsize =(8, 6)) 
  

ax = fig.add_axes([0, 0, 1, 1]) 
  

bp = ax.boxplot(data) 
  

plt.xticks([1,2],['user_1','user_2'])
```

Диаграмма области (Area Chart)

```python
# Area Chart

fig, ax = plt.subplots() #график и оси

#входные данные
x = np.arange(0,100,1)
y = x * 2


ax.fill_between( x, y, color="skyblue", alpha=0.2) #отрисовать область под графиком
ax.plot(x, y, color="Slateblue", alpha=0.6) #отрисовать график
```

Круговая диаграмма (Pie Chart)

```python
# Pie Chart

fig, ax = plt.subplots(figsize=(8,8)) #график и оси

#входные данные
labels = 'Jan', 'Feb', 'March', 'April', 'May', 'June'
user_1 = [10, 3, 15, 21, 17, 14]

ax.pie(user_1, labels=labels, explode=(0.07, 0, 0, 0, 0, 0), autopct='%1.1f%%', startangle=130, shadow=True) #построение
```

График временных рядов (Time Series Plot)

```python
#Time Series Plot
import pandas as pd 

plt.figure(figsize=(8,6)) #график и оси
ts = pd.Series(np.random.randn(100), index = pd.date_range( 
                                '1/1/2020', periods = 100)) #дата серия

ts = ts.cumsum() 
ts.plot() 
```

Поверхность (Surface)

```python
# 3D Surface

import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#график и ось
fig = plt.figure()
ax = fig.gca(projection='3d')

#входные данные
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)#построение
```
